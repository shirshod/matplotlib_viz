### What is Matplotlib?

Matplotlib is an open source, low-level graph plotting library in Python that serves as a visualization utility.
It is mostly written in Python, with a few segments in C, Objective-C and Javascript for platform compatibility.

Matplotlib was created by John D. Hunter. Read more [here](https://github.com/matplotlib/matplotlib). 

### Importing libraries

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Plotting â€” basics

#### Plotting a line with two coordinates

```python
### defines all x-coordinates
xpoints = np.array([0, 6])

### defines all y-coordinates
ypoints = np.array([0, 250])

### plots line (line-plotting is the default when just two points are given)
plt.plot(xpoints, ypoints)
plt.show()
```

#### Plotting just the two points with coordinates

```python
xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

### plots points 
plt.plot(xpoints, ypoints, "o")
plt.show()
```

Note: The "o" is shorthand for "ring"

#### Plotting a line chart with multiple points

```python
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

### plots line 
plt.plot(xpoints, ypoints)
plt.show()
```

#### Plotting points when only one coordinate is given

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints, "o")
plt.show()
```

Note: When only one coordinate is given, the other defaults to 0, 1, 2 and so on.

### Markers

#### Plotting a line and marking each point with a circle 

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = "o")
plt.show()
```

#### Plotting a line and marking each point with an asterisk

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = "*")
plt.show()
```

Note: More shorthands for markers can be found [here](https://www.w3schools.com/python/matplotlib_markers.asp).

#### Plotting a dotted red line 

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, "o:r")
plt.show()
```

Note: More shorthands for line and color can be found [here](https://www.w3schools.com/python/matplotlib_markers.asp).

#### Manipulating marker size

```python
ypoints = np.array([3, 8, 1, 10])

### sets marker size at 20
plt.plot(ypoints, marker = "o", ms = 20)
plt.show()
```

#### Manipulating marker color

```python
ypoints = np.array([3, 8, 1, 10])

### sets marker edge-color as red
plt.plot(ypoints, marker = "o", ms = 20, mec = "r")
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

### sets marker face-color as red
plt.plot(ypoints, marker = "o", ms = 20, mfc = "r")
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

### sets both marker edge and face-color as red
plt.plot(ypoints, marker = "o", ms = 20, mec = "r", mfc = "r")
plt.show()
```

Note: Instead of shorthands, hexidecimal color values can also be used. Refer to the color schema [here](https://www.w3schools.com/colors/colors_hexadecimal.asp).

### Drawing multiple lines

#### Drawing two lines from only one coordinate type

```python
y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])

plt.plot(y1)
plt.plot(y2)

plt.show()
```

### Defining labels

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

### defines label for x-axis
plt.xlabel("Average Pulse")
### defines label for y-axis
plt.ylabel("Calorie Burnage")

plt.show()
```

### Subplots

#### Plotting two charts in same figure horizontally

```python
### plot 1
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

### subplot display has one row (1), two columns (2) and is the first plot (1)
plt.subplot(1, 2, 1)
plt.plot(x,y)

### plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

### subplot display has one row (1), two columns (2) and is the second plot (2)
plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
```

#### Plotting two charts in same figure vertically

```python
### plot 1
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

### subplot display has two rows (2), one column (1) and is the first plot (1)
plt.subplot(2, 1, 1)
plt.plot(x,y)

### plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

### subplot display has two rows (2), one column (2) and is the second plot (2)
plt.subplot(2, 1, 2)
plt.plot(x,y)

plt.show()
```

#### Plotting six charts in figure with two rows and three columns

```python
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 1)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 2)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 3)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 4)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 5)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 6)
plt.plot(x,y)

plt.show()
```

Note: The figures are plotted row-wise. Row 1 has first three figures and row 2 last three.

### Pie charts

```python
y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show() 
```

Note: The pie chart draws one piece (called a wedge) for each value in the array (in this case [35, 25, 25, 15]). By default the plotting of the first wedge starts from the x-axis and moves counterclockwise.
The size of each wedge is determined by comparing the value with all the other values, by using the formula x/sum(x).

#### Adding labels

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show() 
```

Note: The default start angle is at the x-axis, but you can change the start angle by specifying a startangle parameter. The startangle parameter is defined with an angle in degrees, default angle is 0.

#### Manipulating start angles

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

### changes start angle to 90 degrees (first wedge starts at 90)
plt.pie(y, labels = mylabels, startangle = 90)
plt.show() 
```

#### Explode wedge

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show() 
```

#### Adding shadows

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show() 
```

#### Manipulating colors

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["black", "hotpink", "b", "#4CAF50"]

plt.pie(y, labels = mylabels, colors = mycolors)
plt.show() 
```

#### Adding legend

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

### labels adds legend
plt.pie(y, labels = mylabels)
### adds legend title
plt.legend(title = "Four Fruits:")
plt.show() 
```
